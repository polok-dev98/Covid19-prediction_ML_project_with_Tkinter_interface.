```python
import tkinter as tk
from tkinter import ttk, messagebox
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
import pickle
from tkinter import *
from PIL import ImageTk, Image

# create the tkinter window
root = tk.Tk()
root.title("COVID-19 Prediction")
root.geometry("600x500")
root.configure(bg="#728FCE")

label = ttk.Label(root, text="COVID-19 Prediction App :",font=("Arial", 20))
label.grid(row=0, column=0, padx=5, pady=15)
# Create labels
label1 = ttk.Label(root, text="Have Cough?",font=("Arial", 10))
label1.grid(row=3, column=0, padx=5, pady=10)
label2 = ttk.Label(root, text="Have Fever?",font=("Arial", 10))
label2.grid(row=4, column=0, padx=5, pady=10)
label3 = ttk.Label(root, text="Have Sore_throat?",font=("Arial", 10))
label3.grid(row=5, column=0, padx=5, pady=10)
label4 = ttk.Label(root, text="Have shortness of breath?",font=("Arial", 10))
label4.grid(row=6, column=0, padx=5, pady=10)
label5 = ttk.Label(root, text="Headache?",font=("Arial", 10))
label5.grid(row=7, column=0, padx=5, pady=10)
label6 = ttk.Label(root, text="Age 60 and above?",font=("Arial", 10))
label6.grid(row=8, column=0, padx=5, pady=10)
label7 = ttk.Label(root, text="Gender?",font=("Arial", 10))
label7.grid(row=9, column=0, padx=5, pady=10)
label8 = ttk.Label(root, text="Test indication?",font=("Arial", 10))
label8.grid(row=10, column=0, padx=5, pady=10)

# Create comboboxes
combo1 = ttk.Combobox(root, values=["Yes", "No"])
combo1.grid(row=3, column=1, padx=5, pady=5)
combo2 = ttk.Combobox(root, values=["Yes", "No"])
combo2.grid(row=4, column=1, padx=5, pady=5)
combo3 = ttk.Combobox(root, values=["Yes", "No"])
combo3.grid(row=5, column=1, padx=5, pady=5)
combo4 = ttk.Combobox(root, values=["Yes","No"])
combo4.grid(row=6, column=1, padx=5, pady=5)
combo5 = ttk.Combobox(root, values=["Yes", "No"])
combo5.grid(row=7, column=1, padx=5, pady=5)
combo6 = ttk.Combobox(root, values=["Yes", "No"])
combo6.grid(row=8, column=1, padx=5, pady=5)
combo7 = ttk.Combobox(root, values=["Male", "Female"])
combo7.grid(row=9, column=1, padx=5, pady=5)
combo8 = ttk.Combobox(root, values=["Contact with confirmed", "Abroad", "Other"])
combo8.grid(row=10, column=1, padx=5, pady=5)

button = tk.Button(root, text="Predict", command=lambda: predict_cases(),width=8,height=3,font=("Arial", 12))
button.grid(row=11, column=0,padx=0,pady=14)

# define a function to clear all the comboboxes
def clear_combos():
    combo1.set("")
    combo2.set("")
    combo3.set("")
    combo4.set("")
    combo5.set("")
    combo6.set("")
    combo7.set("")
    combo8.set("")

# create the clear button
clear_button = tk.Button(root, text="Clear", command=clear_combos,width=8,height=3,font=("Arial", 12))
clear_button.grid(row=11, column=1,padx=5,pady=14)

model = pickle.load(open('model.pkl','rb'))
pca = pickle.load(open('pca.pkl','rb'))

# function to predict the number of cases
def predict_cases():
    try:
        # get the inputs from the user
        one = combo1.get()
        two = combo2.get()
        three = combo3.get()
        four= combo4.get()
        five = combo5.get()
        six = combo6.get()
        seven = combo7.get()
        eight = combo8.get()
        
        # encode the categorical variables
        if one == "Yes":
            one = 1
        elif one == "No":
            one = 0
        if two == "Yes":
            two = 1
        elif two == "No":
            two = 0
        if three == "Yes":
            three = 1
        elif three == "No":
            three = 0
        if four == "Yes":
            four = 1
        elif four == "No":
            four = 0
        if five == "Yes":
            five = 1
        elif five == "No":
            five = 0
        if six == "Yes":
            six = 1
        elif six == "No":
            six = 0
        if seven == "Male":
            seven = 1
        elif seven == "Female":
            seven = 0
        if eight == "Contact with confirmed":
            eight = 1
        elif eight == "Abroad":
            eight = 2
        else:
            eight=0
            
        # load the trained model and make a prediction
        
        x = pca.transform([[one,two,three,four,five,six,seven,eight]])
        
        prediction = model.predict(x)
        if prediction == 0:
            messagebox.showinfo("Prediction", f"Corona result is Negative")
        else:
            messagebox.showinfo("Prediction", f"Corona result is Positive")
        # show the prediction to the user
        
    except:
        messagebox.showerror("Error", "Error Occurs.")





root.mainloop()
```
